```python
# Install gTTS
!pip install -q gtts

from gtts import gTTS
from IPython.display import Audio, display

try:
    text = input("Enter text to convert: ")

    voice = "speech.mp3"
    language = "en"       #we use "ti" for tel,"hi" for hindi,.....

    tts = gTTS(text=text, lang=language, slow=False)   #here,slow=False is normal speed,slow=True is slow speed
    tts.save(voice)

    print("Audio generated successfully!")

    display(Audio(voice, autoplay=True))

except Exception as e:
    print("Error:", e)
```
